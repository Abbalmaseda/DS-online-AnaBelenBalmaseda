Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center" o investigar como funciona bar_label). Los colores de las barras deben tener un 50% de transparencia y ser: azul, gris, rojo, aceituna y marrón.  

2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Arial", tamaño = 14, peso = "Bold", color azul acero o "steelblue")
3. El scatter de área de agua de un ciudad con su poblacion. Transparencia al 50% y color verde.
4. El boxplot de la altura de las ciudades.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def guardar_figura(nombre):
    fig = plt.gcf()
    fig.savefig(nombre, dpi=300, bbox_inches="tight")
    plt.close(fig)
```

```python
top = df_casa_california.sort_values("population_total",ascending= False).head(5)
```

```python
plt.figure()
colors =["blue","grey","red","olive","brown"]
bars = plt.bar(top["city"], top["population_total"], color = colors , alpha = 0.5)
edgecolor= "black"

plt.title("Ciudades más pobladas de California") 
plt.ylabel("Habitantes miles")

plt.bar_label(bars, labels = top["population_total"],label_type="edge")
plt.xticks(rotation = 0)

plt.show()
guardar_figura("ciudades_pobladas.png")

```

```python
plt.figure(figsize=(8,5))
plt.hist(df_casa_california["area_land_sq_mi"], bins = 50 , color = "steelblue" , edgecolor= "black",label = "Hist Area")
plt.title("Histograma del área total ocupada por las ciudades", fontname="Arial", fontsize=14, fontweight="bold")
plt.xlabel("Superficie km$^2$", fontname="Arial", fontsize=14, fontweight="bold")
plt.ylabel("Numero de ciudades", fontname="Arial", fontsize=14, fontweight="bold")
plt.legend()
plt.show()
guardar_figura("Histograma_del_área_total.png")
```

```python
plt.figure(figsize=(6,4))

plt.scatter(
    df_casa_california["area_water_km2"],
    df_casa_california["population_total"],
    color="green",
    alpha=0.5
)


plt.title("Población por kilómetro cuadrado de agua")
plt.xlabel("Área de agua en la ciudad (km$^2$)")
plt.ylabel("Población")


plt.show()

guardar_figura("Población_por_km.png")
```

```python
plt.figure(figsize=(7,5))

plt.boxplot(
    df_casa_california["elevation_m"].dropna(),
    vert=False,              
    patch_artist=True,     
    boxprops=dict(facecolor="red")  
)

plt.title("Diagrama de caja de la Altitud de las ciudades de California",
          fontsize=14, fontweight="bold")

plt.xlabel("Altitud (m)")
plt.yticks([]) 

plt.grid(alpha=0.3)

plt.show()

guardar_figura("Diagrama_de_Altitud.png")
```
